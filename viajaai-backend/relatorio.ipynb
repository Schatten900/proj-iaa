# Etapa 1. Proposta do Tema e Definição dos Produtos
 O ramo de negócio escolhido para este projeto foi turismo e viagens. O objetivo é desenvolver uma plataforma de recomendação de destinos turísticos personalizados, com base no perfil individual dos usuários, suas preferências e avaliações anteriores. O sistema utiliza técnicas com foco em Filtragem Baseada em Conteúdo (Content-Based Filtering) utilizando TF-IDF para gerar recomendações.
 
 Foram definidos 20 destinos turísticos como “produtos” do sistema, e eles são: Praia de Copacabana, Ski nos Alpes Suíços, Disney World em Orlando, Romance em Paris, Mochilão na Europa, Resort em Maldivas, Ecovila Amazônica, Campos do Jordão, Cancún All Inclusive, Tour em Roma, Fernando de Noronha, Las Vegas, Machu Picchu, Safári na África do Sul, Patagônia na Argentina, Bali, New York city, Gramado, Ilhas Gregas e Jericoacoara. Cada um desses destinos possui 3 características principais, extraídas e armazenadas na base de dados para análise. As características são:
 
     Clima: Quente, Frio, Tropical ou Temperado;
	 Preço: Baixo, Médio, Alto ou Luxo;
	 Companhia Ideal: Sozinho, Casal, Família ou Amigos;
     
Além disso, cada destino é classificado com:

	 Lazer: Romance, Aventura, Relaxamento, Histórico, Cultural, Gastronômico, Ecoturismo
	 Gêneros de viagem: Praia, Piscina, Trilha, Compras, Radical
     
Onde o usuário, dentre essas opções, deve escolher as 3 opções que mais se identifica dentre as opções destacadas acima em cada um dos campos.O padrão de avaliação adotado é uma escala de 1 a 5 estrelas, sendo:

	1 estrela: Pouco relevante/preferido
	5 estrelas: Altamente relevante/preferido
    
Essas avaliações são dadas pelos usuários ao cadastrar preferências e ao avaliar destinos visitados.

# Etapa 2. Organização e Formatação dos Dados
A estrutura de dados do sistema foi desenvolvida utilizando MySQL como banco de dados relacional. O banco foi dividido para facilitar a manutenção e testes:

	init-db.sql: Criação do esquema relacional;
	generos.sql, lazeres.sql, viagens.sql, usuarios.sql: População de dados com 50 usuários e 20 destinos distintos;
	avaliacoes.sql: Contém as 500 avaliações simuladas, respeitando os critérios definidos de preferência dos usuários;

Os dados foram simulados de forma manual, tendo sua lógica baseada em combinações de gostos, buscando garantir diversidade e coerência nas avaliações dos usuários desse banco de treinamento. Isso permitiu a criação de uma matriz de utilidade com pelo menos 500 linhas, fundamental para a geração e validação dessas recomendações.

# Etapa 3. Modelo de Recomendação Baseado em Conteúdo
O modelo de recomendação implementado segue a abordagem de Filtragem baseada em conteúdo (Content-Based Filtering), utilizando o algoritmo TF-IDF (Term Frequency - Inverse Document Frequency) como mecanismo para extração de características e análise de similaridade.

O Modelo é dividido em 4 etapas principais: Inicialmente temos a extração dos Atributos, onde o backend realiza a concatenação textual dos atributos de cada viagem, incluindo clima, faixa de preço, companhia, gêneros e lazeres, que vai servir como base para a vetorização.

Em seguida, passa para a vetorização TF-IDF, utilizando o TfidfVectorizer da biblioteca scikit-learn, cada descrição de viagem é transformada em um vetor numérico. As palavras que ocorrem com maior frequência em uma descrição, mas são raras nas demais, recebem maior peso, permitindo identificar o destino que tem maior relação com essas características.
Após esse passo, ocorre o cálculo de similaridade, onde, nesses vetores é aplicada calcula a média da tabela ViagemAvaliacao, e faz a similaridade por cosseno do perfil do usuário com todas as viagens do sistema e multiplica ambos os dados, sendo 70% da similaridade por cosseno e 30% pela média da tabela, gerando uma matriz de similaridade que orienta a recomendação.

Por fim, temos a geração das recomendações que, com uma solicitação, o sistema faz a recuperação do perfil do usuário, calcula a média ponderada das avaliações de lazer e gênero, compara os vetores e retorna os destinos que apresentam maior similaridade com esse perfil gerado.

# - Representação dos Dados no Backend (Inteiros e Enums)

As categorias de gênero e lazer foram implementadas no backend como valores inteiros, uma decisão que impacta diretamente no funcionamento do nosso sistema de recomendação. E essa escolha foi feita porque esse sistema possui vantagem em relação à aceleração das comparações e vetorização, facilitando o uso dos modelos baseados nesse aprendizado de máquina.

Já a escolha da modelagem com Enums foi utilizada para otimização do processo interno, já que essas representações são ideais para isso.
Cada destino é representado como um vetor numérico multidimensional, onde cada dimensão representa uma categoria de preferência (ex: lazeres e gêneros). Esse vetor é fundamental para identificar proximidade entre o perfil do usuário e os destinos disponíveis.

# Etapa 4. Aplicativo e Interface do Usuário
A interface web do sistema foi projetada para oferecer um fluxo de tela em que inicialmente o usuário se encontra na página de login. Nela, caso o usuário não possua uma conta, pode clicar em “aqui” para ser redirecionado para a tela de cadastro. 

Na tela de cadastro, o usuário deve indicar seu nome, e-mail e senha. E, após a realização do login, deverá preencher um formulário com seu clima preferido, faixa de preço ideal e companhia de viagem desejada.

Em seguida, há a definição de preferências lazer e gênero, onde o usuário deve escolher 3 preferências em cada categoria, e essas escolhas serão avaliadas por ele entre 1 e 5 estrelas, representando a intensidade de sua escolha. Por fim, o usuário poderá acessar uma página com os destinos sugeridos com base na similaridade entre suas preferências e os atributos dos destinos disponíveis, podendo ver detalhes do destino, simular uma compra e avaliar a viagem após essa compra, impactando suas futuras recomendações.

# Integração com Backend:
O frontend envia uma requisição POST com o ID do usuário para o endpoint /recomendacoes. A partir daí o controller aciona o serviço de recomendação que realiza as consultas no banco de dados e aplica o modelo TF-IDF, calcula a similaridade e retorna uma lista de destinos mais compatíveis com o perfil do usuário. Por fim, o frontend exibe as recomendações com base no ranking de similaridade.
